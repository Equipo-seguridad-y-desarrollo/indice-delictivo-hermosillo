```python
# solo de hermosillo
imc_hermosillo = imc2020[imc2020['NOM_MUN'] == 'Hermosillo']
imc_hermosillo
```

```python
import folium
# Asegúrate de tener el CRS correcto
imc_hermosillo = imc_hermosillo.to_crs(epsg=4326)

# Centro del mapa
centro = [imc_hermosillo.geometry.centroid.y.mean(), imc_hermosillo.geometry.centroid.x.mean()]

# Crear mapa base
m = folium.Map(location=centro, zoom_start=11, tiles='cartodb positron')

# Añadir polígonos con tooltip de la colonia
folium.GeoJson(
    imc_hermosillo,
    name="Colonias de Hermosillo",
    tooltip=folium.GeoJsonTooltip(fields=["COLONIA"], aliases=["Colonia:"])
).add_to(m)

# Mostrar mapa
m
```

```python
import os
# lo convertimos en un csv 

# --- 1. Definir la ruta de salida ---
# (Asegúrate de que la carpeta ..\data\raw exista)
output_folder = r"..\data\raw"
output_csv_path = os.path.join(output_folder, 'poligonos_hermosillo.csv')

if 'imc_hermosillo' in locals():
    try:
        # --- 3. Crear una copia para no modificar el original ---
        df_para_csv = imc_hermosillo.copy()
        
        # --- 4. Convertir la geometría a texto (WKT) ---
        # Un CSV no puede guardar el objeto "geometry", así que lo convertimos a texto
        if 'geometry' in df_para_csv.columns:
            print("Convirtiendo la columna 'geometry' a texto (WKT)...")
            df_para_csv['POLIGONO_WKT'] = df_para_csv['geometry'].to_wkt()
            
            # Eliminamos la columna 'geometry' original porque no se puede guardar en CSV
            df_para_csv = df_para_csv.drop(columns=['geometry'])
        
        # --- 5. Guardar el archivo CSV ---
        df_para_csv.to_csv(
            output_csv_path, 
            index=False, 
            encoding='utf-8-sig' # 'utf-8-sig' para que Excel lea bien acentos/ñ
        )
        
        print(f"\n¡Éxito! Archivo guardado en: {output_csv_path}")
        print("El CSV contiene todas las columnas originales y una nueva llamada 'POLIGONO_WKT'.")

    except Exception as e:
        print(f"Ocurrió un error al guardar el archivo: {e}")
else:
    print("Error: La variable 'imc_hermosillo' no se encontró.")
```

```python
import os
import geopandas as gpd
import requests

# URL base del repositorio de luis Moreno
repo_url = "https://github.com/Sonora-en-Datos/ColoniasSonora/raw/main/shapes/INE_Limpio/"
files = [
    "INE_Limpio.shp", "INE_Limpio.dbf", "INE_Limpio.shx", "INE_Limpio.prj"
]
datos_dir =  r"..\data\raw"
os.makedirs(datos_dir, exist_ok=True)

# Descarga de archivos
for fname in files:
    r = requests.get(repo_url + fname)
    with open(os.path.join(datos_dir, fname), "wb") as f:
        f.write(r.content)
    print(f"Descargado: {fname}")

# Leer shapefile y sacar polígonos
shapefile_path = os.path.join(datos_dir, "INE_Limpio.shp")
gdf = gpd.read_file(shapefile_path)

# Extrae solo los polígonos
poligonos = gdf[gdf.geometry.type == 'Polygon']

# Muestra o guarda los datos filtrados
print(poligonos.head())
poligonos.to_file(os.path.join(datos_dir, "poligonos.shp"))

# Para otros archivos de manzanas (mza_ine_limpio.*), repite el procedimiento cambiando los nombres.
```

```python
# colonias de hermosillo 
colonia_hermosillo = poligonos[poligonos['nom_loc'] == 'Hermosillo']
colonia_hermosillo
```

```python
# guardamos en csv las colonias hermosillo en un csv 
output_csv_path_hermosillo = os.path.join(datos_dir, 'poligonos_hermosillo.csv')
colonia_hermosillo.to_csv(output_csv_path_hermosillo, index=False, encoding='utf-8-sig')
print(f"\n¡Éxito! Archivo guardado en: {output_csv_path_hermosillo}")
```
