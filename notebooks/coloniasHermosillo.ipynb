```python
import pandas as pd

# habitantes por colonia en sonora de los datos de luis moreno
url = "https://raw.githubusercontent.com/Sonora-en-Datos/ColoniasSonora/main/resultados/sonora_colonias_2020.csv"

df = pd.read_csv(url)

df.head()
```

```python
# solo las colonias de hermosillo 
df_col = df[df['nom_loc'] == 'Hermosillo']
df_col.head()
```

```python
# ordenar por columna de poblacion
df_col_sorted = df_col.sort_values(by='poblacion_total', ascending=False)
df_col_sorted.head(10)
```

```python
df_col_sorted.info()
# guardarlo localmente
df_col_sorted.to_csv('hermosillo_colonias_poblacion.csv', index=False)
```

```python
#comntar 
colonias = df_col_sorted['nom_col'].tolist()
```

```python
# contamos las colonias 
len(colonias)
```

```python
# revisar si hay repetidas 
data = pd.DataFrame(colonias, columns=['colonia'])
data['count'] = 1   
data = data.groupby('colonia').count().reset_index()
data = data[data['count'] > 1]
data
```
